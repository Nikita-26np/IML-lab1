```python
'''Write a Pandas program to implement following operation
 to convert a NumPy array to a Pandas series
 to convert the first column of a DataFrame as a Series
 to create the mean and standard deviation of the data of a given
Series
 to sort a given Series'''
```

```python
import pandas as pd  
import numpy as np 
```

```python
''' to convert a NumPy array to a Pandas series'''

a1=np.array([1,2,3,4,5])
print("Numpy array : ",a1)
print("converted panda series :")
print(pd.Series(a1))
```

```python
'''to convert the first column of a DataFrame as a Series'''

technologies = {
    'Courses':["java","C++","C","Python","C#"],
    'Fee' :[20000,25000,26000,22000,24000],
    'Duration':['60days','40days','35days','50days','60days'],
    'Discount':[1500,2000,1000,3500,2500]
     }
df = pd.DataFrame(technologies)
print (df)

df2 = df.iloc[:, 0]
print(df2)
```

```python
'''to create the mean and standard deviation of the data of a given Series'''

a = pd.Series(data = [2,5,3,7,1,9,2,6,4,8])
print ("Original Data Series : ",a)
print ("Mean of the Data series : ",a.mean())
print ("Standard deviation of the data series : ",a.std())
```

```python
'''to sort a given Series'''

a = pd.Series([9,8,3,5,1,2])
print ("Original data series : ")
print(a)
new_a = pd.Series(a).sort_values()
print ("Soerted Series : ")
print(new_a)
```
