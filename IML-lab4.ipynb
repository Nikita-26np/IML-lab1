```python
'''Write a NumPy program to implement following operation
 to add, subtract, multiply, divide arguments element-wise
 to round elements of the array to the nearest integer
 to calculate mean across dimension, in a 2D numpy array
 to calculate the difference between neighboring elements, elementwise of a given arra'''
```

```python
import numpy as np
```

```python
'''to add, subtract, multiply, divide arguments element-wise'''
  
a = np.arange(10)
print("\na: \n", a)
print("\nnp.add(a, 2):\n", np.add(a, 2))                 #addition
print("\nnp.subtract(a, 5):\n", np.subtract(a, 5))       #subtraction
print("\nnp.multiply(a, 3):\n", np.multiply(a, 3))       #Multiplication
print("\nnp.divide(a, 2):\n", np.divide(a, 2))           #Division 
```

```python
''' to round elements of the array to the nearest integer'''

x=np.array([1.1,1.2,2.5,3.5,4.9,5.1])
print("Nearest integer of array element ",np.round(x));
```

```python
''' to calculate mean across dimension, in a 2D numpy array'''

a = np.array([[2, 3], [4, 5]])
print("a:\n", a)

print("Mean of a: ", np.mean(a))
print("Mean of a along axis=0: ", np.mean(a, axis=0))
print("Mean of a along axis=1: ", np.mean(a, axis=1))
```

```python
''' to calculate the difference between neighboring elements,-element wise of 
a given array'''

x = np.array([1, 2, 5, 8, 9])
np.diff(x)
```
