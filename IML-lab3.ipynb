```python
'''Write a NumPy program to implement following operation
 to split an array of 14 elements into 3 arrays, each with 2, 4, and 8
elements in the original order
 to stack arrays horizontally (column wise)
'''
```

```python
import numpy as np
```

```python
''' to split an array of 14 elements into 3 arrays, each with 2, 4, and 8
elements in the original order'''

a1=np.arange(14)
print(a1)
print("Array with 2 elements: ", a1[:2])
print("Array with 4 elements: ", a1[2:6])
print("Array with 8 elements: ", a1[6:14])
```

```python
'''to stack arrays horizontally (column wise)'''

a1=np.arange(9)
print ("Vertical Array : ",a1)
new_a1= a1.reshape(-1,1)
print("Horizontal Array a1 : ",new_a1)
```
