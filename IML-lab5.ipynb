```python
'''Write a NumPy program to implement following operation
 to find the maximum and minimum value of a given flattened array
 to compute the mean, standard deviation, and variance of a given
  array along the second axis'''
```

```python
import numpy as np
```

```python
'''to find the maximum and minimum value of a given flattened array'''

a = np.arange(9).reshape(3,3)
print("flattened array",a)

print("maximum value of flattened array : ",np.max(a))
print("minimum value of flattened array : ",np.min(a))





```

```python
'''to compute the mean, standard deviation, and variance of a given
   array along the second axis'''

a = np.arange(9).reshape(3,3)
print(a)
print("Mean of a along axis=1: ", np.mean(a, axis=1))
print("Standard Deviation along axis=1: \n", np.std(a, axis=1))
print("Variance along axis=1: \n", np.var(a, axis=1))
```
