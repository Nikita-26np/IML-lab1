```python
'''Write a NumPy program to implement following operation
 to convert a list of numeric values into a one-dimensional NumPy 
array
 to create a 3x3 matrix with values ranging from 2 to 10
 to append values at the end of an array
 to create another shape from an array without changing its data(3*2 
to 2*3)'''
```

```python
import numpy as np 
```

```python
a=[1,2,3,4]

n1=np.array(a)
print(n1)
    
```

```python
a1=np.arange (2,11)
b=a1.reshape(3,3)
print(b)
```

```python
a2=[1,2,3,4]
print('original array :',a2)

b2=np.append(a2,[5,6])
print('after appending :',b2)


```

```python
a1=np.arange (2,8)

b=a1.reshape(3,2)
print (' shape of a1 3*2 ',b)

c=a1.reshape(2,3)
print (' shape of a1 2*3 ',c)
```
